Correlation Matrix
Is an another way when we are dealing with a huge dataset which has more features and they might many such columns which aren't going to yeild much useful information or we can find using the other features.

```python
plt.figure(figsize=(30,15))
corrMatrix = df.corr()
sns.heatmap(corrMatrix, annot=True)
plt.show()
```

**Sand** has high correlation with **SOM** and **Clay**, hence the data linearity will be affected if it is not dropped.
Features with high correlation are more linearly dependent and hence have almost the same effect on the dependent variable. So, when two features have high correlation, we can drop one of the two features.

```python
# with the following function we can select highly correlated features
# it will remove the first feature that is correlated with anything other feature
def correlation(df, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = df.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

```python
corr_features = correlation(df, 0.9) #we have set the threshold to be 0.9 that is 90%.
len(set(corr_features))
corr_features
```

### Detecting the outliers using IQR and removing them.

```python
Q1 = np.percentile(df['PP7'], 25,interpolation = 'midpoint')
Q3 = np.percentile(df['PP7'], 75,interpolation = 'midpoint')
IQR = Q3 - Q1
print("Old Shape: ", df.shape)
# Upper bound
upper = np.where(df['PP7'] >= (Q3+1.5*IQR))
# Lower bound
lower = np.where(df['PP7'] <= (Q1-1.5*IQR))
''' Removing the Outliers '''
df.drop(upper[0], inplace = True)
df.drop(lower[0], inplace = True)
print("New Shape: ", df.shape)
```

### Which content in the soil is more responsible for the emission of the N2O ?

```python
table = pd.pivot_table(df,index=['Month'],values=['PP2','PP7','N2O'], aggfunc=np.mean).round()
df_table=table.sort_values(by=["N2O"],ascending=False)
df_table
```

```python
df_table.plot(kind='bar',figsize=(15,5))
```

From the bar chart it cleared that the precipitation content that is PP2 in the top 25cm of the soil layer over a period of 2 days has  directly proportional to the N2O emission into the air.

```python
table1 = pd.pivot_table(df,index=['Month'],values=['NH4','NO3','N2O'], aggfunc=np.mean).round()
df_table1=table1.sort_values(by=["N2O"],ascending=False)
df_table1
```

```python
df_table1.plot(kind='bar',figsize=(15,5))
```

From the bar chart it cleared that the amount of NH4 and NO3 content in the top 25cm of the soil layer is directly proportional to the N2O emission into the air.
